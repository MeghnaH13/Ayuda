```python

def recommend_courses_based_on_domain_and_skills(goal_domain, user_skills):
    # Ensure goal_domain is a string
    if not isinstance(goal_domain, str):
        return pd.DataFrame()
    
    # Normalize the input domain to lower case
    goal_domain = goal_domain.lower()
    
    # Filter courses by domain
    filtered_courses_df = courses[
        (courses['Domain_1'].str.contains(goal_domain, case=False, na=False)) | 
        (courses['Domain_2'].str.contains(goal_domain, case=False, na=False)) | 
        (courses['Domain_3'].str.contains(goal_domain, case=False, na=False))
    ]
    
    # Debug print: Check filtered courses
    print("Filtered Courses based on Domain:")
    print(filtered_courses_df.head())
    
    # Check if there are courses available in the desired domain
    if filtered_courses_df.empty:
        return pd.DataFrame()
    
    # Combine the user's skills into a single string
    user_skills_str = ', '.join(user_skills)
    
    # Append user skills to the DataFrame using pd.concat
    user_skills_df = pd.DataFrame({'course_name': ['User Skills'], 'Skills_associated': [user_skills_str]})
    temp_df = pd.concat([filtered_courses_df, user_skills_df], ignore_index=True)
    
    # Ensure Skills_associated column is of string type
    temp_df['Skills_associated'] = temp_df['Skills_associated'].astype(str)
    
    # Debug print: Check concatenated DataFrame
    print("\nDataFrame after adding User Skills:")
    print(temp_df.head())
    
    # Convert skillsets to TF-IDF features
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(temp_df['Skills_associated'])
    
    # Debug print: Check TF-IDF Matrix Shape
    print("\nTF-IDF Matrix Shape:", tfidf_matrix.shape)
    
    # Compute cosine similarity
    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
    
    # Debug print: Check Cosine Similarity Matrix Shape
    print("\nCosine Similarity Matrix Shape:", cosine_sim.shape)
    
    # Get similarity scores for the user's skills with all filtered courses
    sim_scores = list(enumerate(cosine_sim[-1][:-1]))
    
    # Debug print: Check Similarity Scores
    print("\nSimilarity Scores:", sim_scores)
    
    # Sort courses based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the indices of the top 5 courses
    top_indices = [i[0] for i in sim_scores[:5]]
    
    # Debug print: Check Top Indices
    print("\nTop Indices:", top_indices)
    
    # Return the top 5 recommended courses
    return filtered_courses_df.iloc[top_indices]

```

```python
# Get user input for their future goal domain
goal_domain = input("Enter your future goal domain (e.g., Data Science, Web Development): ")

# Get user input for their skills
user_skills = input("Enter your skills separated by commas (e.g., Python, Data Analysis, Machine Learning): ").split(',')

# Get the recommended courses
recommended_courses = recommend_courses_based_on_domain_and_skills(goal_domain, user_skills)

# Display the recommended courses
if not recommended_courses.empty:
    print(f"Recommended Courses for the domain '{goal_domain}' based on your skills:\n")
    print(recommended_courses.to_string(index=False))
else:
    print(f"Sorry, no courses found for the domain: {goal_domain}")
```
